#Visualização básica usando Matplotlib

Vamos começar com algumas visualizações básicas. Sem usar isso também, você pode imprimir um gráfico, mas usando isso você será capaz de fazer alterações na figura como se fosse capaz de redimensionar os gráficos dependendo de sua aparência e também salvar o gráfico. Da mesma forma, a variável *'ax'* aqui pode ser usada para fornecer rótulos aos eixos. Abaixo está um exemplo simples onde passei os dados como um array e os imprimi diretamente como um gráfico

#Gráficos de tendência

Agora, vamos começar a usar alguns dados reais e aprender como construir gráficos interessantes e personalizá-los para torná-los mais intuitivos. Conforme explicado, na maioria dos casos de uso da vida real, os dados exigiriam alguma transformação para torná-los utilizáveis ​​para os gráficos. Abaixo está um exemplo em que usei os dados do Netflix, mas transformei os dados para consolidar o número de filmes e programas de TV por ano. E então eu usei a função *‘plt.subplots ()’*, mas também adicionei alguns detalhes adicionais para tornar o gráfico mais intuitivo e autoexplicativo.



```

n_data = netflix_data.groupby(['release_year', 'type'], as_index=False).show_id.count()
n_data.columns = ['release_year', 'type', 'count']
n_data_pivot = n_data.pivot(index='release_year', columns='type', values='count').reset_index()
n_data_pivot.fillna(0, inplace = True)
n_data_pivot.head()
```

```
#  Declare a chart with one Axis
fig, ax = plt.subplots()  

# Creating some  dummy data
ax.plot(n_data_pivot.release_year, n_data_pivot.Movie, label = 'Movies')
ax.plot(n_data_pivot.release_year, n_data_pivot['TV Show'], label = 'Tv Shows') 
ax.set_ylabel('Number of Movies/TV Shows') 
ax.set_xlabel('Year of Release')
ax.set_title("Trend Chart")
plt.text(x=1970, y=400, s=r'Majority Movies are released after 2000 and Tv Shows mostly after 2010')#, fontsize = 12)
fig.set_size_inches(18.5, 10.5)
plt.grid()
ax.legend()
```



Existem mais algumas personalizações que podem ser feitas no gráfico acima, como a criação de um eixo duplo. No caso acima, não há muita diferença entre o número de filmes e programas de TV, portanto, os dados parecem OK, se houver uma grande diferença entre eles, o gráfico não será muito claro, nesses casos podemos fazer uso eixo duplo para que o atributo com valores menores também seja escalado em linha com o outro.

#Gráficos de barra

Os gráficos de barras nos ajudam a comparar vários valores ao mesmo tempo, plotando-os lado a lado. Existem diferentes tipos de gráficos de barras,

  - Gráfico de Barras Verticais

  - Gráfico de Barras Horizontais

  - Gráfico de Barras Empilhadas


Abaixo está um exemplo de um gráfico de barras, há uma série de personalizações adicionadas a este gráfico. Eles são,

  - Rótulos de eixo e títulos são adicionados
  - O tamanho da fonte foi fornecido
  - O tamanho da figura também é fornecido (o gráfico padrão pareceria muito menor e desordenado)
  - Uma função é usada para gerar e adicionar valores ao topo de cada uma das barras para ajudar os visualizadores a obter os detalhes reais



```
# Data from: https://www.kaggle.com/shivamb/netflix-shows
# With transformations

netflix_data['year_added'] = netflix_data['date_added'].str.slice(start=-4)
n_data_added = netflix_data.groupby(['year_added', 'type'], as_index=False).show_id.count()
n_data_added.columns = ['year_added', 'type', 'count']
n_data_added = n_data_added.pivot(index='year_added', columns='type', values='count').reset_index()
n_data_added.fillna(0, inplace = True)
n_data_added.head()
```

```

labels = n_data_added['year_added']

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
Movies_rects = ax.bar(x - width/2, n_data_added['Movie'], width, label='Movies') # Add color here
TVshows_rects = ax.bar(x + width/2, n_data_added['TV Show'], width, label='TV Show')

# Add some text for labels, title, configure chart size
ax.set_xlabel('Year added on Netflex')
ax.set_ylabel('Count of Number of Movies/TV Shows')
ax.set_title('Number of movies and TV Shows by year they got added')
ax.set_xticks(x)
ax.set_xticklabels(labels)
fig.set_size_inches(18.5, 10.5)
plt.rcParams.update({'font.size': 15})
ax.legend()

 #Function to generate labels on top of the bars
def gen_label(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


gen_label(Movies_rects)
gen_label(TVshows_rects)

fig.tight_layout()
plt.show()
```

#Gráfico de barras horizontais e empilhadas

Os gráficos de barras verticais são mais comuns, mas também podemos fazer uso dos gráficos de barras horizontais, especialmente quando as etiquetas de dados têm um nome longo e é muito difícil imprimi-los abaixo de uma barra vertical. No caso do gráfico de barras empilhadas, as barras serão empilhadas umas sobre as outras dentro de uma categoria. abaixo está o exemplo de implementação dos gráficos de barras horizontais e empilhadas. O código abaixo também inclui personalização para a cor do gráfico.





```

data = pd.DataFrame(dict(Year = n_data_added['year_added'],
                       Movie = n_data_added['Movie'], TVshow=n_data_added['TV Show'])) 

ind = np.arange(len(data))
width = 0.4

fig1, ax1 = plt.subplots()
ax1.barh(ind, data.Movie, width, color='red', label='Movies')
ax1.barh(ind + width, data.TVshow, width, color='green', label='TV Shows')

ax1.set(yticks=ind + width, yticklabels=data.Year, ylim=[2*width - 1, len(data)])

ax1.set_ylabel('Year added on Netflex')
ax1.set_xlabel('Count of Number of Movies/TV Shows')
ax1.set_title('Number of movies and TV Shows by year they got added')

ax1.legend()
fig1.set_size_inches(10, 10)
plt.show()
```

#Gráfico de pizza e donut

Os gráficos de pizza são úteis para mostrar a proporção de diferentes categorias nos dados e esses gráficos de pizza podem ser facilmente modificados para um gráfico de rosca cobrindo a parte central do gráfico de pizza com um círculo e realinhando o texto / valores para se adequar à rosca gráfico. Abaixo está um exemplo simples onde implementei o gráfico de pizza e, posteriormente, o modifiquei em um gráfico de rosca




```
# Small modification to data to suit for the Pie Chart
n_data_added['Total'] = n_data_added['Movie'] + n_data_added['TV Show']
# To select last few rows only
n_data_added_flt = n_data_added[-4:]
n_data_added_flt = n_data_added_flt.reset_index(drop=True) # For resetting index
n_data_added_flt.head()
```

```

# Pie chart
labels = n_data_added_flt['year_added']
sizes = n_data_added_flt['Total']
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice 

fig2, ax2 = plt.subplots()
ax2.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax2.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

#Mapas de calor

Os mapas de calor são outra visualização interessante amplamente usada em dados de série temporal para revelar as sazonalidades e outros padrões no conjunto de dados. No entanto, para construir um mapa de calor, precisamos transformar os dados em um formato específico para suportar a plotagem de mapa de calor. Abaixo está um código de amostra para transformar os dados de acordo com o gráfico do mapa de calor e a biblioteca marítima usada para construir o mapa de calor



```
# Data for heatmap
agg_data2 = heart_failure_data.groupby(['age', 'ejection_fraction'], as_index=False).DEATH_EVENT.sum()
agg_data2.columns = ['Age', 'ejection_fraction', 'Deaths']
agg_data2_p = agg_data2.pivot(index='Age', columns='ejection_fraction', values='Deaths')
agg_data2_p.fillna(0, inplace=True)
agg_data2_p.head(20)
```

Mapa de calor

```

plt.subplots(figsize=(20,15))
sns.heatmap(agg_data2_p, annot=True)
plt.show()
```



#Criação de gráficos interativos

Ao trabalhar em projetos de ciência de dados, às vezes havia a necessidade de compartilhar alguma visualização com as equipes de negócios. Ferramentas de painel são amplamente usados ​​para esse propósito, mas digamos que haja um padrão interessante que você percebeu ao realizar a análise de dados e gostaria de compartilhá-lo com o usuário empresarial. Se eles forem compartilhados como uma imagem, pode não haver muito que o usuário empresarial possa fazer, mas se forem compartilhados como um gráfico interativo, isso dá ao usuário empresarial o poder de ver os detalhes granulares aumentando ou diminuindo o zoom ou usar outra funcionalidade para interagir com o gráfico. Abaixo está um exemplo onde estamos criando um arquivo HTML como uma saída que inclui a visualização que pode ser compartilhada com qualquer outro usuário e pode ser simplesmente aberto em um navegador da web.



```

from bokeh.plotting import figure, output_file, show

p = figure(title = "Age Vs Ejection Fraction")
p.xaxis.axis_label = 'Age'
p.yaxis.axis_label = 'Ejection Fraction'

p.circle(agg_data2["Age"], agg_data2["ejection_fraction"],
         fill_alpha=0.2, size=10)

output_file("test.html", title="Example")

show(p)
```
